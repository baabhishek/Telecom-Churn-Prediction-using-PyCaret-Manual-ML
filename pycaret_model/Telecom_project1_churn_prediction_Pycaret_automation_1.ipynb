```python
import pandas as pd
import numpy as np
```

```python
dataset=pd.read_csv('/content/sample_data/telecomproj1_data_forPycaret.csv')
dataset
```

```python
dataset.columns
```

```python
#for training set
data=dataset.sample(frac=0.8,random_state=42).reset_index(drop=True)
#for test set(rename col for test after frac =0.8)
data_unseen=dataset.drop(data.index).reset_index(drop=True)
```

```python
print(f'training set: {data.shape}\ntest set: {data_unseen.shape}')
```

```python
from pycaret.classification import *
```

```python
classification_model=setup(data=data, target='Churn', normalize=True,normalize_method='zscore', fold_shuffle=True, fold=10, session_id=42)
```

```python
# comapre model performance

compare_models()
```

```python
# knn  giving good accuracy create knn :

knn_og_model=create_model('knn')
```

```python
# hyperparameter tuning /(fine tuning):(original vs tuned--> which one is good?)

knn_tunned_model=tune_model(knn_og_model, optimize='Recall')
```

```python
#evaluate the model -> tunned model is better but not manual jupyter accu recall:

evaluate_model(knn_tunned_model)
```

```python
# predict unseen data

unseen_prediction=predict_model(knn_tunned_model,data=data_unseen)
unseen_prediction.head()
```

```python
# saving the model

save_model(knn_tunned_model,"colabs_telecom_proj1_pycaret")
```
